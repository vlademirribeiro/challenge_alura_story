### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
display(loja)
display(loja2)
display(loja3)
display(loja4)
```

#1. Análise do faturamento


```python
#Calculando faturamento de cada loja
faturamento_total = loja['Preço'].sum()
faturamento_total2 = loja2['Preço'].sum()
faturamento_total3 = loja3['Preço'].sum()
faturamento_total4 = loja4['Preço'].sum()

print(f'faturamento: R$ {faturamento_total:,.2f}')
print(f'faturamento: R$ {faturamento_total2:,.2f}')
print(f'faturamento: R$ {faturamento_total3:,.2f}')
print(f'faturamento: R$ {faturamento_total4:,.2f}')


```

```python
import matplotlib.pyplot as plt

def gerar_grafico(lojas):
    """
    Gera um gráfico de barras com o faturamento total de cada loja.

    Parâmetros:
    lojas (list): Lista de tuplas com o DataFrame da loja e o nome da loja.
    """
    total_lojas = []
    faturamento = []

    for loja, nome_loja in lojas:
        total_lojas.append(nome_loja)
        faturamento.append(loja['Preço'].sum())

    plt.figure(figsize=(10, 6))
    plt.bar(total_lojas, faturamento, color= plt.cm.Pastel2_r.colors)
    plt.title('Faturamento Total por loja', fontsize= 12)
    plt.xlabel('Lojas', fontsize=12)
    plt.ylabel('Faturamento', fontsize=12)
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.savefig('faturamento_total_loja.png', dpi=200, bbox_inches='tight')

    plt.show()


lojas = [(loja, 'Loja 1'), (loja2, 'Loja 2'), (loja3, 'Loja 3'), (loja4, 'Loja 4')]
gerar_grafico(lojas)
```

# 2. Vendas por Categoria


```python
# Agrupando as vendas por cat de produto em cada loja
cat_loja = loja.groupby('Categoria do Produto')['Preço'].sum()
cat_loja2 = loja2.groupby('Categoria do Produto')['Preço'].sum()
cat_loja3 = loja3.groupby('Categoria do Produto')['Preço'].sum()
cat_loja4 = loja4.groupby('Categoria do Produto')['Preço'].sum()

# Juntando os dados em um único DataFrame
vendas_cat = pd.DataFrame({
    'Loja ': cat_loja,
    'Loja 2': cat_loja2,
    'Loja 3': cat_loja3,
    'Loja 4': cat_loja4
})

# Substituindo NaNs por 0
vendas_cat.fillna(0, inplace=True)

# Somando as vendas totais por cat
vendas_cat['Total'] = vendas_cat.sum(axis=1)
```

```python
print("Vendas por categoria de produto em cada loja:\n")
vendas_cat.sort_values(by='Total', ascending=False)
```

# 3. Média de Avaliação das Lojas

```python
media_avaliacao_por_loja = pd.DataFrame({
    'Lojas': ['Loja 1','Loja 2','Loja 3','Loja 4'],
    'Media Avalc. Compra': [
        loja["Avaliação da compra"].mean(),
        loja2["Avaliação da compra"].mean(),
        loja3["Avaliação da compra"].mean(),
        loja4["Avaliação da compra"].mean()
    ]
})
```

```python
print("Avaliação média por loja (de 1 a 5):")
media_avaliacao_por_loja.sort_values(by='Media Avalc. Compra', ascending=False).round(2)
```

# 4. Produtos Mais e Menos Vendidos

```python

def mais_vendidos_por_loja(df, loja_nome):
    mais_vendidos = df['Produto'].value_counts().head().reset_index()
    mais_vendidos.columns = ['Produto', 'Qtd']
    mais_vendidos.columns = pd.MultiIndex.from_product([[loja_nome], mais_vendidos.columns])
    return mais_vendidos

# Dicionário com os DataFrames e nomes das lojas
lojas = {
    'Loja 1': loja,
    'Loja 2': loja2,
    'Loja 3': loja3,
    'Loja 4': loja4,
}

# Gerar os mais vendidos de cada loja
mais_vendidos_list = [mais_vendidos_por_loja(df, nome) for nome, df in lojas.items()]

# Concatenar lado a lado
mais_vendidos_juntas = pd.concat(mais_vendidos_list, axis=1)

print(f"Produtos Mais Vendidos por Loja:")
display(mais_vendidos_juntas)
```

```python
def menos_vendidos_por_loja(df, loja_nome):
    menos_vendidos = df['Produto'].value_counts().tail().reset_index()
    menos_vendidos = menos_vendidos.rename_axis('Produto')
    menos_vendidos.columns = ['Produto', 'Qtd']
    menos_vendidos.columns = pd.MultiIndex.from_product([[loja_nome], menos_vendidos.columns])
    return menos_vendidos

# Dicionário com os DataFrames e nomes das lojas
lojas = {
    'Loja 1': loja,
    'Loja 2': loja2,
    'Loja 3': loja3,
    'Loja 4': loja4,
}

# Gerar os menos vendidos de cada loja
menos_vendidos_list = [menos_vendidos_por_loja(df, nome) for nome, df in lojas.items()]

# Concatenar lado a lado
menos_vendidos_juntas = pd.concat(menos_vendidos_list, axis=1)

print(f'Produtos Menos Vendidos por Loja:')
display(menos_vendidos_juntas)
```

# 5. Frete Médio por Loja


```python
import pandas as pd

# Supondo que loja, loja2, loja3, loja4 são DataFrames definidos anteriormente
lista_lojas = [loja, loja2, loja3, loja4]
nomes_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

df_frete = pd.DataFrame([
    {
        'Loja': nome,
        'Média Frete': df['Frete'].mean(),
        'Total Frete': df['Frete'].sum()
    }
    for nome, df in zip(nomes_lojas, lista_lojas)
])

print(df_frete)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_frete['Total Frete'], df_frete['Média Frete'], color=['#8A2BE2', '#00FFFF', '#8A2BE2', '#7FFFD4'], s=150)


for i in range(len(df_frete)):
    plt.text(df_frete['Total Frete'][i],
             df_frete['Média Frete'][i] +.2,
             df_frete['Loja'][i],
             fontsize=12,
             ha='right',
             va='center')

plt.xlabel('Total de Frete (R$)')
plt.ylabel('Média de Frete (R$)')
plt.title('Dispersão: Média x Total de Frete por Loja')
plt.grid(True)
plt.tight_layout()
plt.ylim(31, 35.5)
plt.xlim(72000,82000)
plt.show()
```
